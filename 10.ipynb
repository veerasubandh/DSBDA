```python
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
setosa_df = df[df['species'] == 'setosa']
versicolor_df = df[df['species'] == 'versicolor']
virginica_df = df[df['species'] == 'virginica']
```

```python
print("\nSubset: Setosa")
setosa_df.head()
```

```python
print("\nSubset: Versicolor")
versicolor_df.head()
```

```python
print("\nSubset: Virginica")
virginica_df.head()
```

```python
merged_df = pd.concat([setosa_df, versicolor_df])
print("\nMerged Subset (Setosa + Versicolor):")
merged_df
```

```python
sorted_df = df.sort_values(by='petal_length', ascending=True)
print("\nSorted Data by Petal Length:")
print(sorted_df[['species', 'petal_length']])
```

```python
transposed_df = df.transpose()
print("\nTransposed Data (Top 3 rows):")
transposed_df
```

```python
melted_df = pd.melt(df, id_vars=['species'], value_vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
print("\nMelted Data (Long Format):")
melted_df
```

```python
pivot_df = df.pivot_table(values='petal_length', index='species', aggfunc='mean')
print("\nPivoted (Wide Format) by Petal Length Mean:")
print(pivot_df)
```
