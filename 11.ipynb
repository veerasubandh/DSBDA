```python
import pandas as pd
```

```python
df = pd.read_csv("tmdb_5000_movies.csv")
df
```

```python
english_movies = df[df['original_language'] == 'en']
english_movies
```

```python
french_movies = df[df['original_language'] == 'fr']
french_movies
```

```python
merged_movies = pd.concat([english_movies, french_movies])
print("\nMerged Movies (English + French):")
merged_movies
```

```python
sorted_movies = df.sort_values(by='vote_average', ascending=False)
print("\nTop Movies Sorted by Customer Ratings:")
sorted_movies[['title', 'vote_average']]
```

```python
transposed_data = df.transpose()
print("\nTransposed Data:")
transposed_data
```

```python
melted_df = pd.melt(df,
                    id_vars=['title', 'original_language'],
                    value_vars=['budget', 'revenue', 'vote_average'])

print("\nMelted Data (Long Format):")
melted_df
```

```python
pivot_df = df.pivot_table(values='vote_average', index='original_language', aggfunc='mean')
print("\nPivoted Data (Wide Format) - Avg. Vote by Language:")
pivot_df
```
