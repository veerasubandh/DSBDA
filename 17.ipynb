```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
plt.figure(figsize=(10, 6))
df['age'].hist(bins=20, color='pink', edgecolor='black')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.stripplot(x='target', y='chol', data=df, jitter=True, palette='Set2')
plt.title("Dot Plot of Cholesterol by Heart Disease Presence")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Cholesterol")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='age', data=df,palette='Set2')
plt.title("Bar Plot of Average Age by Sex")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Average Age")
plt.show()
```

```python
age_hr = df.groupby('age')['thalach'].mean().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(age_hr['age'], age_hr['thalach'], marker='o')
plt.title("Line Chart of Max Heart Rate vs Age")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate (thalach)")
plt.grid(True)
plt.show()
```

```python
import seaborn.objects as so  # seaborn 0.12+ required

# Boxplot with histogram for 'chol'
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['chol'], color='orange')
sns.histplot(df['chol'], bins=30, kde=True, color='skyblue', alpha=0.5)
plt.title("Box Plot with Histogram - Cholesterol")
plt.xlabel("Cholesterol")
plt.show()

# Boxplot with scatter overlay for 'age'
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='age', data=df, palette='Set2')
sns.stripplot(x='target', y='age', data=df, jitter=True, color='blue', alpha=0.4)
plt.title("Box Plot with Scatter Overlay - Age vs Heart Disease")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```
