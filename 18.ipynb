```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heart.csv") 
df
```

```python
plt.figure(figsize=(8, 5))
df['age'].hist(bins=15, color='teal', edgecolor='black')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(False)
plt.show()
```

```python
# Pie chart for target variable (heart disease)
labels = ['No Disease (0)', 'Disease (1)']
sizes = df['target'].value_counts().sort_index()

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['skyblue', 'coral'])
plt.title("Heart Disease Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y='chol', data=df, color='orange')
plt.title("Box Plot of Cholesterol")
plt.ylabel("Cholesterol (mg/dl)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='thalach', hue='target', palette='Set1')
plt.title("Age vs Max Heart Rate (thalach)")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.legend(title="Heart Disease")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='thalach', data=df, palette='Set2')
sns.stripplot(x='target', y='thalach', data=df, jitter=True, color='black', alpha=0.5)
plt.title("Max Heart Rate vs Heart Disease (Box + Scatter)")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Max Heart Rate")
plt.grid(True)
plt.show()
```
