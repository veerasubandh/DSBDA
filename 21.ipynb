```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("adult.csv")
df
```

```python
df.replace('?', pd.NA, inplace=True)
df.dropna(inplace=True)
```

```python
plt.figure(figsize=(8, 5))
df['age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(x='income', y='educational-num', data=df, jitter=True, palette='pastel')
plt.title("Dot Plot of Education Num vs Income")
plt.xlabel("Income")
plt.ylabel("Education Level (Numeric)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='education', hue='income', palette='Set2')
plt.xticks(rotation=90)
plt.title("Education vs Income")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.legend(title="Income")
plt.show()
```

```python
avg_hours = df.groupby('age')['hours-per-week'].mean()

plt.figure(figsize=(10, 5))
plt.plot(avg_hours.index, avg_hours.values, color='blue')
plt.title("Average Working Hours by Age")
plt.xlabel("Age")
plt.ylabel("Average Hours per Week")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Boxplot
sns.boxplot(x='income', y='hours-per-week', data=df, palette='coolwarm')

# Scatter (dot) overlay
sns.stripplot(x='income', y='hours-per-week', data=df, color='black', alpha=0.3, jitter=True)

plt.title("Working Hours vs Income (Box + Dot Plot)")
plt.xlabel("Income Category")
plt.ylabel("Hours per Week")
plt.grid(True)
plt.show()
```
