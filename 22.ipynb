```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("adult.csv")
df
```

```python
df.shape
```

```python
df.replace("?", pd.NA, inplace=True)
df.dropna(inplace=True)
```

```python
df.isnull().sum()
df.dropna
```

```python
plt.figure(figsize=(10, 5))
df['age'].plot(kind='hist', bins=30, color='teal', edgecolor='black')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
income_counts = df['income'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(income_counts, labels=income_counts.index, autopct='%1.1f%%', colors=['coral', 'skyblue'])
plt.title("Income Distribution Pie Chart")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='income', y='hours-per-week', data=df, palette='coolwarm')
plt.title("Box Plot of Hours per Week by Income Category")
plt.xlabel("Income")
plt.ylabel("Hours per Week")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='hours-per-week', hue='income', data=df, alpha=0.6)
plt.title("Scatter Plot of Age vs Hours per Week")
plt.xlabel("Age")
plt.ylabel("Hours per Week")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='income', y='hours-per-week', data=df, whis=1.5, palette='pastel')
sns.stripplot(x='income', y='hours-per-week', data=df, color='black', alpha=0.3, jitter=True)
plt.title("Boxplot + Scatter of Hours per Week by Income")
plt.xlabel("Income")
plt.ylabel("Hours per Week")
plt.grid(True)
plt.show()
```
