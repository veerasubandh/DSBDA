```python
import pandas as pd
```

```python
df = pd.read_csv("hepatitis.csv")
df
```

```python
for col in df.columns:
    try:
        df[col] = pd.to_numeric(df[col])
    except:
        pass  # Keep original if conversion fails
```

```python
male_df = df[df['sex'] == 1]   # Male
female_df = df[df['sex'] == 2] # Female
```

```python
merged_df = pd.concat([male_df, female_df])
merged_df
```

```python
sorted_df = df.sort_values(by=['age', 'sgot', 'protime'], ascending=True)
sorted_df
```

```python
transposed_df = df.transpose()  # Transpose first 10 rows for readability
transposed_df
```

```python
melted_df = pd.melt(df, id_vars=['class', 'sex'], 
                    value_vars=['age', 'sgot', 'protime'],
                    var_name='Measurement', value_name='Value')
melted_df
```

```python
casted_df = melted_df.pivot_table(index=['class', 'sex'], 
                                  columns='Measurement', 
                                  values='Value', 
                                  aggfunc='mean').reset_index()
casted_df
```
