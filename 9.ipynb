```python
import pandas as pd 
```

```python
df = pd.read_csv('dataset_Facebook.csv', sep=';') #loading the dataset 
df.head()
```

```python
status_df = df[df['Type'] == 'Status'] #creating a subset status
photo_df = df[df['Type'] == 'Photo']  #creating a subset photo
```

```python
print("\nSubset: Status Posts")
status_df.head()
print("\nSubset: Photo Posts")
photo_df.head()
```

```python
merged_df = pd.concat([status_df, photo_df]) #merging two subsets
print("\nMerged Subset (Status + Photo):")
merged_df
```

```python
sorted_df = df.sort_values(by='Page total likes', ascending=False) #Sort Data on Page total likes
print("\nSorted Data by Page Total Likes:")
print(sorted_df[['Page total likes', 'Type']].head())
```

```python
transposed_df = df.transpose() #transposing the data
print("\nTransposed Data:")
transposed_df
```

```python
melted_df = pd.melt(df,
                    id_vars=['Type', 'Category'],
                    value_vars=['Page total likes', 'Post Month', 'Post Weekday'])  #Melting Data to long format 
print("\nMelted Data (Long Format):")
melted_df
```

```python
pivot_df = df.pivot_table(values='Page total likes', index='Category', columns='Type', aggfunc='mean')
print("\nPivoted (Wide Format):")
pivot_df
```
